The purpose of this notebook is to quantify bacterial cells in a set of environmental samples. As usual, we can start by importing the packages we need for the analysis.

```python
import os
import FlowCytometryTools as fct
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

Next, we import data from the calibration experiment. This was an experiment where cultured cells were grown, fixed, diluted across a range of concentrations, stained, and counted. The first file is a spreadsheet containing the expected cell concentrations for each dilution based on colony counts from a companion plating experiment, as well as some other information. Based on the snippet of the spreadsheet printed, we can see that some controls were run including (1) just counting beads (2) fresh media (3) incubated, uninoculated media. We will use the former to ascertain gating parameter for the beads and the latter to quantify sources of noise. 

```python
exp_cnt_file = "/Volumes/KeithSSD/RockCreek/RC_CellCounts/data/SYBR_PI_Dilution_Config.txt"
exp_cnt_df_pre = pd.read_csv(exp_cnt_file, sep="\t", index_col=0)

exp_cnt_df_pre.sort_values(by='Exp. Cell Density')
```

The log of the expected counts are plotted here. We can see that they cover quite with the smallest at 40 and the largest at 7.5e8 cells per mL. We don't expect to recover this range, but we do want to see where we can come close

We define a function here to pull out some of the metadata for each sample including start and end times, as well as PMT voltage gains. This will give us a sense of what order samples were processed in and if cytometer settings were changed in a way that may affect how events are selected. 

```python
def pull_relavent_metadata(x):
    meta = x.get_meta()
    _channels_ = meta['_channels_'].copy()
    voltages = _channels_[_channels_['$PnN'].str.endswith("-A")].set_index('$PnN')['$PnV'].to_dict()
    voltages = {i[:-2]+' gain':j for i, j in voltages.items()}
    voltages['startTime'] = meta['$BTIM']
    voltages['endTime'] = meta['$ETIM']
    return voltages
```

Here we read all the data files into a dictionary of dataframes. We shift all the values in each column to ensure all values are positive and non-zero. You can see based on the output, that each sample has a different number of events.

```python
data_dir = "/Volumes/KeithSSD/RockCreek/RC_CellCounts/data/SYBR_PI_Dilutions"
data_mags, meta_datas = {}, {}
for i in os.listdir(data_dir):
    if i.endswith(".fcs"):
        key = int(i.split("_Tube_")[1][:-4])
        data_obj = fct.FCMeasurement(ID=key, datafile=os.path.join(data_dir, i))
        meta_datas[key] = pull_relavent_metadata(data_obj)
        data_mags[key] = data_obj
```

```python
other_channels = ['Time']
measured_channels = ["FSC-A", "SSC-A", "488 530/30-A", "488 585/42-A", "488 670LP-A",
                    "488 780/60-A", "633 660/20-A", "633 780/60-A"]
b_value = 400
data_mats = {}
for key, data_obj in data_mags.items():
    temp_df1 = data_obj.data[measured_channels + other_channels]
    temp_df2 = temp_df1.copy()
    for ch_ in measured_channels:
        temp_df2[ch_] = data_obj.transform('hlog', channels=[ch_], b=b_value).data[ch_]
        if int(key) == 6:
            min_, q1_, med_, q3_, max_ = np.percentile(temp_df1[ch_].values, [0, 25, 50, 75, 100])
            print("Pre-transform {}\t{:.2f}\t{:.2f}\t{:.2f}\t{:.2f}\t{:.2f}".format(ch_, min_, q1_, med_, q3_, max_))
            min_, q1_, med_, q3_, max_ = np.percentile(temp_df2[ch_].values, [0, 25, 50, 75, 100])
            print("Post {}\t{:.2f}\t{:.2f}\t{:.2f}\t{:.2f}\t{:.2f}".format(ch_, min_, q1_, med_, q3_, max_))
    data_mats[key] = temp_df2.copy()
```

This block combines the two metadata dataframes and then checks for groups of samples with different PMT setings. Then the following block simply prints out all the data in the table so you can see which samples will be different and also pick out which tube number(s) you want to use as negative controls. 

```python
meta_df = pd.DataFrame(meta_datas).T.sort_index()
meta_df.index.name = 'Tube'
exp_cnt_df = pd.concat((exp_cnt_df_pre, meta_df), axis=1, sort=True, verify_integrity=True)
mad_gains = [i for i in exp_cnt_df.columns if 'gain' in i]
exp_cnt_df['all_gains'] = exp_cnt_df[mad_gains].apply(tuple, axis=1)
num_gains = exp_cnt_df['all_gains'].unique().shape[0]
print("There are {} sets of gain settings".format(num_gains))
if num_gains > 1:
    settings_n, settings_c = np.unique(exp_cnt_df['all_gains'].values, return_counts=True)
    for setn, setc in zip(settings_n, settings_c):
        print("\t{} samples have this setting: {}".format(setc, list(zip(mad_gains, setn))))
        print("\n\t{}".format(exp_cnt_df.index[exp_cnt_df['all_gains'] == setn]))

exp_cnt_df.drop('all_gains', inplace=True, axis=1)
```

The samples used for selecting the gating limits for beads and for noise are defined in the next block. We can do this with PCA, which should allow for separation to be maximized. We list the sames we will use as negatives and list the cutoff for PCA to define populations and then the info is enumerated, printed, and displayed as a simple scatter plot, as well as two distinct contour plots with kernel density estimations. 

```python
#negative_controls = [36, 28, 1, 2, 3, 4]
negative_controls = [2, 3, 4]
PC_bead_threshold = 4

bead_df = pd.concat([data_mats[i].copy() for i in negative_controls], axis=0, ignore_index=True)
real_cols = [i for i in bead_df.columns if i != 'Time']

x = StandardScaler().fit_transform(bead_df[real_cols].values)
pca = PCA(n_components=2, random_state=0)
jb_pca = pca.fit(x)
principalDf = pd.DataFrame(data = jb_pca.transform(x), columns = ['PC1', 'PC2'])

subset1 = list(principalDf.index[principalDf['PC1'] > PC_bead_threshold])
subset2 = list(principalDf.index[principalDf['PC1'] <= PC_bead_threshold])

plt.close('all'), plt.clf()
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)

targets = [subset1, subset2]
labels = ['Subset1', 'Subset2']
colors = ['r', 'g']

ax.scatter(principalDf.loc[:, 'PC1'], principalDf.loc[:, 'PC2'], c='k', label='Not Sure', s = 50)

for target, color, lab in zip(targets, colors, labels):
    _ = ax.scatter(principalDf.loc[target, 'PC1'],
               principalDf.loc[target, 'PC2'], 
               c=color, label=lab, s = 50)

_ =ax.legend()
print("Beads: {}, Noise: {}".format(len(subset1), len(subset2)))
```

The first task is to get seperation between the noise and the beads in order to do some gating. We can do this with PCA, which should allow for separation to be maximized. 

```python
def filter_w_thresh(df, b_t, minmax, thresh):
    proc_order = list(b_t.index)
    index_sets = []
    mat_ = [df[po].values.reshape(-1,1) for po in proc_order]
    for col_i, col in enumerate(mat_):
        index_sets.append(( col >= b_t.loc[proc_order[col_i], minmax[0]] ) & ( col <= b_t.loc[proc_order[col_i], minmax[1]] ))
    
    bmat_ = np.hstack(index_sets).astype(int).sum(1)
    return list(df.index[bmat_ >= bmat_.max() - thresh])
```

```python
test_pack = [16, 8, 9 ,18, 12, 20, 24, 26, 36]

bnc16_df = pd.concat([data_mats[i].iloc[500:3000, :] for i in test_pack], ignore_index=True)
x = StandardScaler().fit_transform(bnc16_df[real_cols].values)
pca2 = PCA(n_components=2, random_state=0)
bnc_pca = pca2.fit(x)
bnc_pc2_df = pd.DataFrame(data = bnc_pca.transform(x), columns = ['PC1', 'PC2'])
print(bnc_pc2_df.shape)

from sklearn import cluster

clf = cluster.DBSCAN(eps=0.25, min_samples=100, metric='l1', n_jobs=-1)
clust_labels = clf.fit(bnc_pc2_df.values).labels_
print(set(clust_labels))
targets = [bnc_pc2_df.index[clust_labels == i] for i in [0,1,2]]
print([len(i) for i in targets])
labels = ['0', '1', '2']
colors = ['g', 'r', 'c']
alphas = [1, 0.3, 1.0]

plt.close('all'), plt.clf()
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)
ax.scatter(bnc_pc2_df.loc[:, 'PC1'], 
           bnc_pc2_df.loc[:, 'PC2'], 
           c='k', label='other', s=50)

for target, color, lab, alpha_ in zip(targets, colors, labels, alphas):
    ax.scatter(bnc_pc2_df.loc[target, 'PC1'],
               bnc_pc2_df.loc[target, 'PC2'], 
               c=color, label=lab, s = 50, alpha=alpha_)
ax.legend()
ax.grid()

plt.show()

print("Please identify populations as 0, 1, or 2: ")
cell_sig = int(input("\tCells: "))
noise_sig = int(input("\tNoise: "))
beads_sig = int(input("\tBeads: "))

cell_targets = list(bnc_pc2_df.index[clust_labels == cell_sig])
noise_targets = list(bnc_pc2_df.index[clust_labels == noise_sig])
beads_targets = list(bnc_pc2_df.index[clust_labels == beads_sig])

noise_lim = bnc16_df.loc[noise_targets, real_cols].describe([0.0, 0.99]).T.iloc[:, [-4,-2]]
cell_lim = bnc16_df.loc[cell_targets, real_cols].describe([0.02, 0.99]).T.iloc[:, [-4,-2]]
bead_lim = bnc16_df.loc[beads_targets, real_cols].describe([0.00, 1.0]).T.iloc[:, [-4,-2]]
noise_lim.columns = ['Noise_min', 'Noise_max']
bead_lim.columns = ['Bead_min', 'Bead_max']
cell_lim.columns = ['Cell_min', 'Cell_max']

ss_mix = pd.concat((noise_lim, cell_lim, bead_lim), sort=True, axis=1)
```

```python
fixative_dilution = 0.5 # mL added to the original sample 
bead_concentration = 1e6 # beads per mL
bead_volume = 0.05 # mL
cells_volume = 0.5 # mL 
ratio_multiplier = ((bead_concentration*bead_volume)/(cells_volume))/fixative_dilution

all_count_df = exp_cnt_df.copy()
new_columns = ['runtime', 'beads', 'cells', 'noise', 'concentration', 'noise_rate']
for nc in new_columns:
    all_count_df[nc] = pd.Series(index=all_count_df.index)

for tube_n, bnci_df in data_mats.items():
    beadi_events = filter_w_thresh(bnci_df.copy(), bead_lim, ['Bead_min', 'Bead_max'], 0)
    noisei_events = filter_w_thresh(bnci_df.copy(), noise_lim, ['Noise_min', 'Noise_max'], 0)
    celli_events = filter_w_thresh(bnci_df.copy(), cell_lim, ['Cell_min', 'Cell_max'], 0)
    
    all_count_df.loc[tube_n, ['beads', 'cells', 'noise']] = [len(beadi_events), 
                                                             len(set(celli_events) - set(beadi_events) - set(noisei_events)), 
                                                             len(noisei_events)]
    all_count_df.loc[tube_n, 'runtime'] = (bnci_df.Time.max() - bnci_df.Time.min())/100
    print(all_count_df.loc[tube_n, :].values)
# cells/bead * beads/mL = cells/mL
all_count_df['concentration'] = (all_count_df['cells'] /(all_count_df['beads']+1))*ratio_multiplier
all_count_df['noise_rate'] = all_count_df['noise']/ all_count_df['runtime']
```

```python
clean_counts['log-exp'] = pd.Series(np.log10(clean_counts['Exp. Cell Density'].values + 1), index=clean_counts.index)
clean_counts['log-conc'] = pd.Series(np.log10(clean_counts['concentration'].values + 1), index=clean_counts.index)
exp_cnts = clean_counts['Exp. Cell Density'].unique()

# plot log-exp (x) against log-conc (y)
# sybr gets green, pi gets red
# f / uf pairs get numbers 

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,8), num=1, dpi=100)

for ec_i, ec in enumerate(sorted(list(exp_cnts))):
    ec_ii = ec_i + 1
    print(ec_ii, ec)
    b1_ = (clean_counts['Exp. Cell Density'] == ec) & (clean_counts['Filtered?'] == True)
    b2_ = (clean_counts['Exp. Cell Density'] == ec) & (clean_counts['Filtered?'] == False)
    for boolx, markval in zip([b1_, b2_], ["${}f$".format(ec_ii), "${}u$".format(ec_ii)]):
        if boolx.sum() > 0:
            slice_ = clean_counts.loc[boolx, ['log-exp', 'log-conc', 'PI?']].values
            for row_ in slice_:
                if row_[-1] == True:
                    ax.scatter(row_[0], row_[1], s=160, c='r', marker=markval)
                else:
                    ax.scatter(row_[0], row_[1], s=160, c='g', marker=markval)

ax.set(xlabel='Expected Counts (log10)', ylabel='Observed Counts (log10)');

plt.show()

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

row_selector = (clean_counts['Filtered?'] == True) & (clean_counts['PI?'] == False)
xy = clean_counts.loc[row_selector, ['log-conc', 'log-exp']].values
x, y = xy[:, 0].reshape(-1, 1), xy[:, 1].reshape(-1, 1)
# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(x, y)

# Make predictions using the testing set
y_pred = regr.predict(x)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean absolute error: %.2f' % mean_absolute_error(y, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f' % r2_score(y, y_pred))
```
